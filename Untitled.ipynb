```python
import numpy as np
import matplotlib.pyplot as plt

# --- 1. PROBLEM TANIMLAMALARI ---
# Amaç: y = 4x1 + 3x2 - 0.5x1x2
def fitness_function(x1, x2):
    return 4*x1 + 3*x2 - 0.5*x1*x2

# Kısıt Kontrolü (Ceza Yöntemi)
def check_constraints(x1, x2):
    # Kısıt 1: x1 + x2 <= 8
    if (x1 + x2) > 8:
        return False
    # Kısıt 2: x2 >= 1.5
    if x2 < 1.5:
        return False
    # Sınırlar: 2 <= x1 <= 6 ve 1 <= x2 <= 4
    if not (2 <= x1 <= 6): return False
    if not (1 <= x2 <= 4): return False
    
    return True

# --- 2. GENETİK ALGORİTMA PARAMETRELERİ ---
POPULATION_SIZE = 20   # Popülasyondaki birey sayısı
GENERATIONS = 50       # Kaç nesil dönecek
MUTATION_RATE = 0.1    # Mutasyon ihtimali (%10)

# --- 3. ALGORİTMA FONKSİYONLARI ---

def create_individual():
    """Rastgele bir birey (x1, x2) oluşturur."""
    while True:
        x1 = np.random.uniform(2, 6)
        x2 = np.random.uniform(1, 4)
        if check_constraints(x1, x2):
            return [x1, x2]

def create_population():
    """İlk popülasyonu oluşturur."""
    return [create_individual() for _ in range(POPULATION_SIZE)]

def selection(population):
    """Turnuva yöntemiyle en iyi ebeveynleri seçer."""
    parents = []
    for _ in range(POPULATION_SIZE):
        # Rastgele 3 kişi seç, kavga ettir, en iyisini al
        candidates = [population[np.random.randint(0, len(population))] for _ in range(3)]
        best_candidate = max(candidates, key=lambda p: fitness_function(p[0], p[1]))
        parents.append(best_candidate)
    return parents

def crossover(parent1, parent2):
    """İki ebeveynden genleri karıştırarak çocuk yapar."""
    # Ağırlıklı ortalama ile çaprazlama (Arithmetic Crossover)
    alpha = np.random.random()
    child1_x1 = alpha * parent1[0] + (1-alpha) * parent2[0]
    child1_x2 = alpha * parent1[1] + (1-alpha) * parent2[1]
    
    # Eğer kısıtları sağlamazsa ebeveynlerden birini aynen al
    if check_constraints(child1_x1, child1_x2):
        return [child1_x1, child1_x2]
    else:
        return parent1

def mutation(individual):
    """Bireyde rastgele değişiklik yapar."""
    if np.random.random() < MUTATION_RATE:
        original = individual.copy()
        # x1 veya x2'ye küçük bir sayı ekle/çıkar
        idx = np.random.randint(0, 2)
        individual[idx] += np.random.uniform(-0.5, 0.5)
        
        # Eğer mutasyon sonucu kısıt bozulursa, mutasyonu iptal et
        if not check_constraints(individual[0], individual[1]):
            return original
    return individual

# --- 4. ANA DÖNGÜ (MAIN LOOP) ---
population = create_population()
best_history = []

print("Genetik Algoritma Başlıyor...")

for gen in range(GENERATIONS):
    # 1. Seçilim (Selection)
    parents = selection(population)
    
    # 2. Çaprazlama (Crossover)
    offspring = []
    for i in range(0, POPULATION_SIZE, 2):
        parent1 = parents[i]
        parent2 = parents[i+1]
        child1 = crossover(parent1, parent2)
        child2 = crossover(parent2, parent1) # Tersiyle ikinci çocuk
        offspring.extend([child1, child2])
        
    # 3. Mutasyon (Mutation)
    population = [mutation(ind) for ind in offspring]
    
    # En iyiyi kaydet
    best_ind = max(population, key=lambda p: fitness_function(p[0], p[1]))
    best_score = fitness_function(best_ind[0], best_ind[1])
    best_history.append(best_score)
    
    # Her 10 nesilde bir bilgi ver
    if gen % 10 == 0:
        print(f"Nesil {gen}: En İyi Puan = {best_score:.4f}")

# --- 5. SONUÇLARI GÖSTERME ---
best_solution = max(population, key=lambda p: fitness_function(p[0], p[1]))
final_score = fitness_function(best_solution[0], best_solution[1])

print("\n--- OPTİMİZASYON TAMAMLANDI ---")
print(f"Optimal Raf Yüksekliği (x1): {best_solution[0]:.2f} m")
print(f"Optimal Raf Derinliği (x2): {best_solution[1]:.2f} m")
print(f"Maksimum Verim Puanı: {final_score:.2f}")

# Gelişim Grafiği
plt.plot(best_history)
plt.title('Genetik Algoritma Gelişim Grafiği')
plt.xlabel('Nesil (Generation)')
plt.ylabel('En İyi Fitness Değeri')
plt.grid(True)
plt.show()
```
